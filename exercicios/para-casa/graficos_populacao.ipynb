```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
df_populacao = pd.read_csv("population.csv")
df_populacao
```

```python
df_populacao.shape
```

```python
df_populacao.columns
```

```python
df_sorted = df_populacao.sort_values(by='Population_2020 ', ascending=False)

df_top15 = df_sorted.head(15)

plt.figure(figsize=(10, 5))
plt.bar(df_top15['Country '], df_top15['Population_2020 '], color='violet', edgecolor='blue')

plt.xlabel('Países')
plt.ylabel('População')
plt.title('População dos paises')
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

```python
df_sorted = df_populacao.sort_values(by='Urban_Pop ', ascending=False)

df_top15 = df_sorted.head(15)

plt.figure(figsize=(10, 7))  

plt.barh(df_top15['Country '], df_top15['Urban_Pop '], color='skyblue', edgecolor='blue')
plt.xlabel('População')
plt.ylabel('Países')
plt.title('Percentual da população urbana')

plt.tight_layout()
plt.show()
```

```python
df_sorted = df_populacao.sort_values(by='Migrants_net', ascending=False)
df_top15 = df_sorted.head(15)
plt.figure(figsize=(10, 7))  

plt.barh(df_top15['Country '], df_top15['Migrants_net'], color='green', edgecolor='blue')
plt.xlabel('População')
plt.ylabel('Países')
plt.title('Fluxo de migrantes')

plt.tight_layout()
plt.show()
```

```python
df_sorted = df_populacao.sort_values(by=' World_Share', ascending=False)

df_top15 = df_sorted.head(15)

plt.figure(figsize=(10, 5))
plt.bar(df_top15['Country '], df_top15[' World_Share'], color='red', edgecolor='blue')

plt.xlabel('Países')
plt.ylabel('População')
plt.title('Distribuição Global dos paises')
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```
